```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import Imputer
from sklearn import preprocessing
```

```python

train_data = pd.read_csv('../input/train.csv')
test_data = pd.read_csv('../input/test.csv')

train_data.isnull().sum()
```

```python
na = train_data.isna().sum()/len(train_data)
na
```

```python
na[na>.50]
```

```python
train_data = train_data.drop(columns=na[na>.50].index)
test_data = test_data.drop(columns=na[na>.50].index)
```

```python
train_data.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
from sklearn import preprocessing
from sklearn.metrics import mean_squared_log_error
import numpy as np
```

```python
label_encoder = preprocessing.LabelEncoder()

obj = train_data.select_dtypes(include='object').columns
obj
for i in obj:
    
    train_data[i] = train_data[i].fillna('text')
    train_data[i] = train_data[i].astype(str)
    test_data[i] = test_data[i].fillna('aaaaaaaa')
    test_data[i] = test_data[i].astype(str)
    
    train_data[i]= label_encoder.fit_transform(train_data[i])
    test_data[i]= label_encoder.fit_transform(test_data[i])
```

```python
my_imputer = Imputer()

train_data = pd.DataFrame(my_imputer.fit_transform(train_data),columns=list(train_data.columns))
test_data = pd.DataFrame(my_imputer.fit_transform(test_data),columns=list(test_data.columns))
```

```python
y = train_data.SalePrice
X = train_data.drop(['SalePrice'], axis=1)
X.head()
```

```python
## splitting the data for training and testing and cleaning it using imputation 

X_train, X_test, y_train, y_test = train_test_split(X.as_matrix(), y.as_matrix(), test_size=0.25)
```

```python
regr = RidgeCV()
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)

print("Test RMSE log = %.4f" % np.sqrt(mean_squared_log_error(y_test, y_pred)))
```

```python
y_pred = regr.predict(test_data)
```

```python
ans = pd.DataFrame({'id':range(1461,1461+len(test_data)),'SalePrice':y_pred})
ans.to_csv('submission.csv',index=False)
```
